### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
# code here
football = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/dataset-session-19/Fifa Worldcup 2022 - Sheet1.csv')
````

````python
football.info()
````

````python
def Znormalization(group,col_per):
  for col in col_per:
    group[col] = (group[col]-group[col].mean())/(group[col].std())
    return group

col_pal = ['Passes','Passes Completed','Attempted Line Breaks','Completed Line Breaks']
groups = football.groupby('Team',group_keys=False).apply(Znormalization,col_pal)
groups.groupby('Team').agg({
    'Passes':['min','max'],
    'Yellow Cards':['min','max','mean'],
    'Attempted Line Breaks':['min','max','std'],
    'Possession (%)':'mean'
})
````

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
# code here
df=pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/dataset-session-19/ipl_deliveries - ipl_deliveries.csv')
df.columns
````

````python
chasing_matches = df[df['innings'] == 2]

# Highest score while chasing
highest_scores = chasing_matches.groupby('batter')['batsman_run'].max()
highest_score_batsman = highest_scores.idxmax()

# Best Strike rate while chasing and have faced 100+ balls
balls_faced = chasing_matches.groupby('batter')['bowler'].count()
qualified_batsmen = balls_faced[balls_faced >= 100].index
strike_rates = (chasing_matches.groupby('batter').sum()['batsman_run'] / balls_faced) * 100
best_strike_rate_batsman = strike_rates[qualified_batsmen].idxmax()

# Chasing mean team batting in second inning
mean_team_batting_score = chasing_matches.groupby('ID')['total_run'].sum().mean()

print("Batsman with highest score while chasing:", highest_score_batsman)
print("Batsman with the best strike rate while chasing and faced 100+ balls:", best_strike_rate_batsman)
print("Mean team batting score in the second inning while chasing:", mean_team_batting_score)
````
