###`Q-8:` Team1 vs Team2 Dashbord
Create a function which will take two string(name of two teams) as input. Show win Loss record between them and player getting most player of the match award in matches between these two teams.
```
team1_vs_team2('Kolkata Knight Riders','Chennai Super Kings')
```

````python
# code here
def team1_vs_team2(team1,team2):
  team_match = ipl[((ipl['Team1'] == team1)&(ipl['Team2'] == team2))|((ipl['Team1'] == team2)&(ipl['Team2']==team2))]
  team1_wins = team_match[team_match['WinningTeam'] == team1].shape[0]
  team2_wins = team_match[team_match['WinningTeam'] == team2].shape[0]
  print(f"{team1} vs {team2} Win_loss Record:")
  print(f"{team1}:{team1_wins} wins, {team2}:{team2_wins} wins")
  player_awards = team_match['Player_of_Match'].value_counts()
  top_player = player_awards.idxmax()
  most_awards = player_awards.max()
  print(f"Player with most Player-of-the-Match awards: {top_player} ({most_awards} awards)")

team1_vs_team2("Kolkata Knight Riders", "Chennai Super Kings")
````

### `Q-10:` Find out the average margin for the team Mumbai Indians for only the session 2011.

````python
# code here

# Assuming you have already loaded the dataset into a DataFrame named 'ipl'
ipl_m = ipl[(ipl['Season'] == '2011') & ((ipl['Team1'] == 'Mumbai Indians') | (ipl['Team2'] == 'Mumbai Indians'))]
# Calculate the margin for each match
ipl_m['Margin'] = abs(ipl_m.apply(lambda row: row['Margin'] if row['WonBy'] == 'Wickets' else -row['Margin'], axis=1))
# Calculate the average margin
average_margin = ipl_m['Margin'].mean()
print(f"Average margin of victory for Mumbai Indians in the 2011 IPL season: {average_margin:.2f} runs")
````
