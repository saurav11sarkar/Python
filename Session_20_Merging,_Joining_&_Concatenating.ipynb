```python
import numpy as np
import pandas as pd
```

```python
cdf = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/datasets-session-20/courses.csv')
ddf = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/datasets-session-20/deliveries.csv')
mdf = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/datasets-session-20/matches.csv')
nadf = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/datasets-session-20/reg-month1.csv')
dedf = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/datasets-session-20/reg-month2.csv')
sdf = pd.read_csv('/content/drive/MyDrive/Compasx/Data Science Mentorship Program/dataset/datasets-session-20/students.csv')
```

```python
# pd.concat
# ignore_index
# df.append
# mullitindex -> fetch using iloc
# concat dataframes horizontally
```

```python
# pd.concat
regs = pd.concat([nadf,dedf],ignore_index=True)
regs
```

```python
# df.append
nadf.append(dedf,ignore_index=True)
```

```python
multi = pd.concat([nadf,dedf],keys=['Nov','Dec'])
multi.loc['Dec']
```

```python
multi = pd.concat([nadf,dedf],keys=['Nov','Dec'])
multi.loc[('Dec',4)]
```

```python
pd.concat([nadf,dedf],axis=1)
```

```python
# inner join
sdf.merge(regs,how='inner',on='student_id')
```

```python
# left join
cdf.merge(regs,how='left',on='course_id')
```

```python
temp = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'parner':[28,26,27]
})
temp
```

```python
student = pd.concat([sdf,temp],ignore_index=True)
```

```python
student.tail()
```

```python
student.merge(regs,how='right',on='student_id')
```

```python
# outer join
student.merge(regs,how='outer',on='student_id').tail(10)
```
