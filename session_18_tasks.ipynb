````python
# code here
teem =df['Team'].unique().tolist() + df['Against'].unique().tolist()
teem = sorted(list(set(teem)))
team_per = {}
for i in teem:
  football_teem = df[(df['Team']==i) | (df['Against']==i)]
  Total_Attempts = football_teem['Total Attempts'].sum()
  On_Target = football_teem['On Target'].sum()

  percentage_attemt = (On_Target/Total_Attempts)*100 if Total_Attempts != 0 else 0
  team_per[i] = round(percentage_attemt,2)

team_per = dict(sorted(team_per.items(),key=lambda item:item[1],reverse=True))

team_per
````

### `Q-2:` Find out how many times the teams are played in this Fifa Worldcup-2022. On top of this, find out the ranks of the teams.

Note: The `DataFrame.rank()` method takes an optiinal parameter named `method`. This parameter takes different values, but one of them is `average` which is by-default. So, when you do the rank, you will get some 2.5 like floating values. But if you change the value as `first`, then you will get in integers but the datatype will be float. So, try with `method="first"` parameter.

````python
# code here
df['Team'].shape[0]
matches_played = pd.DataFrame()
matches_played['Matches_Played'] = (df['Team'].value_counts() + df['Against'].value_counts()).sort_values()
matches_played.reset_index(inplace=True)
matches_played.columns=['Team','Matches_Played']
matches_played['Rank'] = matches_played['Matches_Played'].rank(method='first',ascending=False)
matches_played.fillna(matches_played.mean(),inplace=True)
matches_played['Rank'] = matches_played['Rank'].astype(int)

matches_played.sort_values(by='Rank',inplace=True)
matches_played
````

###`Q-8:` Player with most appearance in final match.

`Team1Players` and `Team2Players` have all players name. It is not a list of players name instead it is str. So handle it as string.

Hint: split and strip will help; Make a series of all players in final and do value counts


````python
# code here
Team1Players = ipl[ipl['MatchNumber'] == 'Final']['Team1Players']
Team2Players = ipl[ipl['MatchNumber'] == 'Final']['Team2Players']
all_player = pd.Series(Team1Players+Team2Players)
player = all_player.apply(lambda x: x.split(',')).explode()
player = player.str.strip()
player_count = player.value_counts()
most_final = player_count.sort_values(ascending=False).index[0]
print(f"Player with the most appearances in the final : {most_final}")
````

````python
# Extract all players from Team1 and Team2 columns
players = pd.concat([ipl[ipl['MatchNumber'] == 'Final']['Team1Players'], ipl[ipl['MatchNumber'] == 'Final']['Team2Players']])

# Split the string into a list of players
players = players.apply(lambda x: x.split(",")).explode()

# Strip any leading and trailing whitespace from each player name
players = players.str.strip()

# Create a Series of all players and count their occurrences
player_counts = players.value_counts()

# Get the player with the most appearances in the final
most_final_appearances_player = player_counts.sort_values(ascending=False).index[0]

# Print the player with the most appearances in the final
print(f"Player with the most appearances in the final: {most_final_appearances_player}")
````
